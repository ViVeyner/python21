# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Общие требования к программе:

 - код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
 - в коде отсутствуют глобальные переменные (за исключением documents и directories);
 - пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды “q”).
 
 ### Задание 1
#### Пункт 1. 
Пользователь по команде “p” может узнать владельца документа по его номеру
#### Пункт 2. 
Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится
#### Пункт 3. 
Пользователь по команде “l” может увидеть полную информацию по всем документам
#### Пункт 4. 
Пользователь по команде “ads” может добавить новую полку
#### Пункт 5. 
Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def master_name():
    """
    Функция master_name() возвращает имя владельца документа по введенному номеру документа

    """
    enter_num = str(input('Введите номер документа:'))
    prove = 0

    for doc in documents:
        if doc['number'] == enter_num:
            name = doc['name']
            print(f'Владелец документа: {name}')
            prove += 1
    if prove == 0:
        print('Документ не найден в базе')
```

```python
def shelf():
    """
    Функция shelf() возвращает по номеру документа на какой полке он хранится

    """
    enter_doc_num = str(input('Введите номер документа:'))
    prove = 0

    for shelf in directories:
        for number in directories[shelf]:
            if number == enter_doc_num:
                print('Документ хранится на полке: ', shelf)
                prove += 1
    if prove == 0:
        print('Документ не найден в базе')
```

```python
def full():
    """
    Функция full() возвращает полную информацию по всем документам, перебирая документы и находя полку, на которой он хранится

    """
    for doc in documents:
        for shelf in directories:
            for number in directories[shelf]:
                if number == doc['number']:
                    print('№: ', doc['number'], ',',
                          'тип: ', doc['type'], ',',
                          'владелец: ', doc['name'], ',',
                          'полка хранения: ', shelf)
```

```python
def add_new_shelf():
    """
    Функция add_new_shelf() добавлет новую полку при условии, что такой полки ранее не было

    """

    shelf_num = str(input('Введите номер полки: '))

    if shelf_num in directories.keys():
        print('Такая полка уже существует. ')
    else:
        directories[shelf_num] = []
        print('Полка добавлена. ')

    all_shelfs = ', '.join(list(directories.keys()))
    print(f'Текущий перечень полок: {all_shelfs}')
```

```python
def del_shelf():
    """
    Функция del_shelf() удаляет полку при условии, что она существует и пуста

    """

    shelf_num = str(input('Введите номер полки: '))

    if shelf_num not in directories.keys():
        print('Такой полки не существует. ')
    elif directories[shelf_num] == []:
        del(directories[shelf_num])
        print('Полка удалена. ')
    else:
        print('На полке есть документ, удалите их перед удалением полки. ')

    all_shelfs = ', '.join(list(directories.keys()))
    print(f'Текущий перечень полок: {all_shelfs}')
```

```python
def prog_call(pr_name):
    """
    Функция prog_call(pr_name) вызывает в работу функцию, переданную в параметрах pr_name.
    Доступные варианты pr_name:

        p - вызывает функицию master_name()
        s - вызывает функицию shelf()
        l - вызывает функицию full()
        ads - вызывает функицию add_new_shelf()
        ds - вызывает функицию del_shelf()
    """

    if pr_name == 'p':
        master_name()
    elif pr_name == 's':
        shelf()
    elif pr_name == 'l':
        full()
    elif pr_name == 'ads':
        add_new_shelf()
    elif pr_name == 'ds':
        del_shelf()
    else:
        print('Данной команды не найдено. Введите другую команду')
```

```python
def main():
    print('Работа программы начата. Введите q для завершения')
    enter = str(input('Введите команду '))

    while enter != 'q':
        prog_call(enter)
        enter = str(input('Введите команду: '))
    print('Работа программы завершена')
```

```python
main()
```
