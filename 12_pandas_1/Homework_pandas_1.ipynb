### Задание 1

Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

Выбран датасет на 1M записей.

```python
import pandas as pd
```

```python
ratings = pd.read_table('movie_rate/ratings.dat', sep='::', names=['UserID', 'MovieID', 'Rating', 'Timestamp'], engine='python')
ratings.tail()
```

```python
movies = pd.read_table('movie_rate/movies.dat', sep='::', names=['MovieID', 'Title', 'Genres'], engine='python')
movies.head()
```

```python
best_movie = (ratings[ ratings['Rating'] == 5 ]['MovieID'].value_counts().head(1)
.rename_axis('MovieID').reset_index(name='Counts of 5'))

#пока не понятно до конца по лекциям как сделать через groupby,действую объясненным методом
```

```python
best_movie.merge(movies, on = 'MovieID', how = 'left')
```

### Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
power[ ((power.category == 4) | (power.category == 12) | (power.category == 21)) 
      & (power.year >= 2005) & (power.year <= 2010)
      & (power.quantity >= 0) 
      & ((power.country == 'Latvia') | (power.country == 'Lithuania') | (power.country == 'Estonia')) ]['quantity'].sum()
```
